The `sort()` method sorts the list **in-place** (modifies the original list directly) and returns `None`.

The `sorted()` function returns a **new sorted list** and leaves the original list unchanged.

```
my_list = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"Original list: {my_list}")

# Using sort() method
list_copy_sort = my_list.copy() # Create a copy to show in-place modification
print(f"List before sort(): {list_copy_sort}")
return_value_sort = list_copy_sort.sort()
print(f"List after sort(): {list_copy_sort}")
print(f"Return value of sort(): {return_value_sort}")

print("\n---")

# Using sorted() function
list_copy_sorted = my_list.copy() # Create a copy to show original list remains unchanged
print(f"List before sorted(): {list_copy_sorted}")
new_sorted_list = sorted(list_copy_sorted)
print(f"New list from sorted(): {new_sorted_list}")
print(f"Original list after sorted() function: {list_copy_sorted}")

```

An **f-string** (formatted string literal) provides a concise and readable way to embed Python expressions inside string literals for formatting. You prefix the string with `f` or `F` and then enclose expressions within curly braces `{}` inside the string. The expressions are evaluated at runtime and their values are inserted into the string.

```
name = "Alice"
age = 30
occupation = "Software Engineer"

# Using an f-string to combine variables into a string
print(f"Hello, my name is {name}. I am {age} years old and I work as a {occupation}.")

# You can also perform operations directly inside the curly braces
radius = 5
area = 3.14159 * (radius ** 2)
print(f"The area of a circle with radius {radius} is {area:.2f}.") # Using .2f for 2 decimal places

# Or call functions
print(f"The uppercase name is {name.upper()}.")
```

The syntax `:{specifier}` inside an f-string is used for formatting the value of the expression before the colon.

In `area:.2f`:
- `area` is the variable whose value you want to format.
- `:` introduces the format specifier.
- `.2f` is the format specifier itself:
  - `.` (dot) indicates that the number should be displayed with a decimal point.
  - `2` specifies the precision, meaning exactly two digits should appear after the decimal point.
  - `f` indicates that the number should be presented as a fixed-point number (a floating-point number).

If the number has fewer than two decimal places, it will be padded with zeros. If it has more, it will be rounded to two decimal places.

```
value1 = 123.45678
value2 = 7.0
value3 = 100

print(f"Value 1 with .2f: {value1:.14f}") # Rounds to two decimal places
print(f"Value 2 with .2f: {value2:.2f}") # Adds two decimal places
print(f"Value 3 with .2f: {value3:.32f}") # Adds two decimal places
```

### Object Identity: `is` and `id()`

*   **`id(object)`**: Returns a unique integer (the memory address) for the specified object. Every object has a unique ID during its lifetime.
*   **`is` operator**: Checks if two variables point to the **same object** in memory (i.e., `id(a) == id(b)`).
*   **`==` operator**: Checks for **equality**—whether the values of two variables are the same, even if they are stored in different memory locations.

```
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"ID of a: {id(a)}")
print(f"ID of b: {id(b)}")
print(f"ID of c: {id(c)}")

print("\nComparisons:")
print(f"a == b: {a == b} (Same values)")
print(f"a is b: {a is b} (Different memory addresses)")
print(f"a is c: {a is c} (Same memory address/reference)")
```

### Creating Independent Copies

To prevent changes in one list from affecting another, you must create a new object in memory.

1.  **`.copy()` method**: Creates a shallow copy.
2.  **Slicing `[:]`**: An idiomatic way to copy the entire list.
3.  **`list()` constructor**: Passing the old list into the constructor.

```
l1 = [1, 2, 3, 4, 5]

# Create a copy instead of a reference
l2 = l1.copy()
# OR l2 = l1[:]

l2.append(6)

print(f"l1: {l1} (Remains unchanged)")
print(f"l2: {l2} (Only this one has 6)")
print(f"l1 is l2: {l1 is l2} (They are different objects)")
```

### Shallow Copy vs. Deep Copy

*   **Shallow Copy**: Creates a new collection object, but the elements inside are still references to the original objects. Use `.copy()` or `[:]`.
*   **Deep Copy**: Creates a completely independent clone of the original object and all objects nested within it. Requires the `copy` module.

```
import copy

# A nested list (list inside a list)
original = [[1, 2, 3], [4, 5, 6]]

# 1. Shallow Copy
shallow = original.copy()

# 2. Deep Copy
deep = copy.deepcopy(original)

# Modify an inner element
original[0][0] = "CHANGED"

print(f"Original: {original}")
print(f"Shallow:  {shallow}  <-- Inner list also changed!")
print(f"Deep:     {deep}  <-- Completely independent!")
```

**Slicing `[:]`**: An idiomatic way to copy the entire list.

```
list1=[1,2,3,4,5]
list2=list1[:]
print(list1 is list2)
print(id(list1),id(list2))
list1.append(6)
print(list1,list2)
```
